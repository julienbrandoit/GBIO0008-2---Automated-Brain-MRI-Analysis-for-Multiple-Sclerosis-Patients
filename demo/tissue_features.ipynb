## Peak test for unimodality

This section aims to give some insights on how we use the peak test method as a test for unimodality and bimodality of a lesion. We apply the test on several lesions to show how the classification is made

```python
from scipy.signal import find_peaks
import numpy as np

lesion_number = [0,8]

for index in lesion_number:
    data = np.array(quantitative_list_007[index])

    plt.figure(figsize=(10, 6))
    pic_number = []
    images = ['FLAIR', 'MT', 'PD', 'R1', 'R2s']
    for dim in range(1, data.shape[1]):
        plt.subplot(2, 2, dim)
        sns.histplot(data[:, dim], bins='auto', kde=True)

        ax = plt.gca() 
        line = ax.lines[0] 
        x_kde, y_kde = line.get_data()
        
        indices_pics, _ = find_peaks(y_kde, height=0.05 * np.max(y_kde),prominence=0.05) 
        pic_number.append(len(indices_pics))
        plt.title(f'Histogram for Dimension {images[dim]}')

        plt.plot(x_kde[indices_pics], y_kde[indices_pics], 'ro', markersize=8, label='Peaks')

    plt.suptitle(f'Histogram for Lesion {index +1}', fontsize=15)
    plt.tight_layout()
    plt.show()



    if np.sum(pic_number) > (data.shape[1]-1):
        print('Bimodal distribution')

    else:
        print('Unimodal distribution')
```

The findpeaks method is based on a unidimensional approach, meaning we apply the unimodality test independently to each dimension and analyze each variable separately.

One example is shown with lesion 1, where the peak test has identified its distribution as unimodal. Indeed, we can see that each dimension presents one peak, identified by the red dot.

Another example showing a bimodal distribution is lesion 9. We can see that in the MT and R2* dimensions, there are two peaks, indicating the bimodal distribution of the lesion.

## Clustering test for unimodality

This section aims to give some insights on how we use the OPTICS automatic clustering method as a test for unimodality or multimodality of a lesion. We apply the clustering method on the 3th lesion which is classified as multimodal by the clustering test but not by the peaks test. 

```python
index = 29
part2.lesion_quantities_histogram(quantitative_list_007[index],index+1)

print("Is the lesion unimodal according to the peaks test ? (1 is Yes, 0 is No):", part2.find_peaks_fct(quantitative_list_007[index], index))
print("Is the lesion unimodal according to the clustering test ? (1 is Yes, 0 is No):", 1 if part2.modality_using_clustering(quantitative_list_007[index])[0] == 1 else 0)
```

We can effectivly see that the two tests don't agree. The peaks test is only based on the marginal distributions but the clustering test 'sees' more information that cannot be capture by the marginals. We can try to visualize the cluster to better understand the result of the clustering test. To do so, we first cluster the points using the OPTICS method and then plot the cluster in the PCA plan of the two principal components of the data. We remove 'noisy' data (detected by optics) to have a better visualisation. 

```python
import numpy as np
from sklearn.cluster import OPTICS
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = np.array(quantitative_list_007[index])


# Standardize the features
X = StandardScaler().fit_transform(data[:, 1:])

# Apply OPTICS clustering
clustering = OPTICS(min_samples=10).fit(X)
labels = clustering.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

centroids = []
for cluster_id in range(n_clusters_):
    cluster_points = data[labels == cluster_id][:, 1:]
    centroid = np.mean(cluster_points, axis=0)
    centroids.append(centroid)

# Filter out noisy data (labels == -1)
filtered_data = data[labels != -1]
filtered_labels = labels[labels != -1]

# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(filtered_data[:, 1:])

# Print the PCA components (directions)
print("PCA Components (directions):")
print(pca.components_)

# Plotting with cluster labels as colors
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=filtered_labels, s=50, cmap='viridis', alpha=0.5)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.title(f'2D PCA Plot of Clustering Result: {n_clusters_} clusters')
plt.savefig('clustering_result.png')
plt.show()
```

One can see that there is indeed two groups of points (two clusters) that seems to have different values of PCA : the first cluster is more on the left plan while the second cluster is more on the right plan. Since the direction of the PCA is not parallel to one of the feature direction, it is not obvious on the marginals and therefore undetected by the peak tests but well detected by the clustering one.

### On the impact of the threshold value

The threshold determine the trade-off sensitivity-specifity. The user can choose the value of the threshold to match their needs. In practice, decreasing the value (with a minimum of 2 observations per cluster) increase the probability of a lesion to be classified as multimodal.

We can highlight this fact by showing the number of clusters as a function of the threshold for example on the lesion number 30.

```python
index = 29

data = np.array(quantitative_list_007[index])

# Standardize the features
X = StandardScaler().fit_transform(data[:, 1:])
n_cluster_list = []
for k in range(2, 25):
    # Apply OPTICS clustering
    clustering = OPTICS(min_samples=k).fit(X)
    labels = clustering.labels_
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_cluster_list.append(n_clusters_)

plt.figure(figsize=(10, 6))
plt.plot(range(2, 25), n_cluster_list, marker='o')
plt.xlabel('Minimum Samples')
plt.ylabel('Number of Clusters')
plt.title('Number of Clusters vs. Minimum Samples')
plt.grid()
plt.savefig('optics_min_samples.png')
plt.show()
```

## Comparison

This section aims to provide a comparison of the two methods used as unimodality tests. One method is multidimensional, while the other is unidimensional, and differences between the two methods could therefore arise. A first step involves examining the total number of unimodal distributions identified using the two methods for a given patient, such as patient 007.

```python
print("Number of unimodal distributions using peak test for patient '007':", patient_007_data['Unimodal distribution peaks test'].sum())
```

```python
print("Number of unimodal distributions using clustering for patient '007':", patient_007_data['Unimodal distribution clustering'].sum())
```

It can be seen that the total number of unimodal distributions is exactly the same using the two methods. However, it may be interesting to delve into the details by examining whether the two methods return unimodality of the distribution for the same lesions of a given patient.

```python
patient_007_data[['Lesion number','Unimodal distribution peaks test', 'Unimodal distribution clustering']]
```

Even if both methods return the same total number of unimodal distributions, some lesions may be identified as unimodal in one test and bimodal in the other test. This highlights differences between the two methods, indicating that using one method over the other will affect the results obtained. An example where the two methods do not agree on the unimodality of a lesion is shown below. The lesion used is the 1st.

```python
index = 0
part2.lesion_quantities_histogram(quantitative_list_007[index],index+1)

print("Unimodal distribution according to the peaks test: ",bool(part2.find_peaks_fct(quantitative_list_007[index], index)))
print("Unimodal distribution according to the clustering test: ", True if part2.modality_using_clustering(quantitative_list_007[index])[0] == 1 else False)
```

For this lesion, it can be seen that the clustering method predicts a unimodal distribution, while the peak test predicts a bimodal distribution. In conclusion, both approaches can detect bimodality but yield different results. This means that caution must be exercised when using the data obtained. However, as the peak test is based on the marginal distribution, it may not capture all the information present in the data.
